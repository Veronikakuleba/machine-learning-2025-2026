Mamy tekst CSV zapisany jako zwykły string.

Waszym celem jest:

rozdzielić tekst na linie,
wyciągnąć nagłówki,
zbudować listę słowników products,
policzyć średnią cenę produktów droższych niż 10.

```python

# Tekst CSV zapisany jako zwykły string
csv_text = """product,price
apple,8
banana,12
milk,15
bread,9
cheese,20
"""
# TODO:
# Rozdziel tekst na linie i zapisz wynik do zmiennej lines

lines = csv_text.split("\n")

print(lines)


# TODO:
# Pierwsza linia zawiera nagłówki kolumn
# Rozdziel ją po przecinku i zapisz do zmiennej headers

headers = lines[0].split(",")
print(headers)


# Tworzymy pustą listę na produkty
products = []
# TODO:
# Przejdź po liniach od drugiej linijki (bez nagłówka)
# i wypisz każdą linię

for line in lines[1:]:
    print(line)

# Dla każdej linii:
# - pomiń pustą linię
# - rozdziel dane po przecinku
# - wypisz values

for line in lines[1:]:
    if line == "":
        continue

    values = line.split(",")
    print(values)


lines = csv_text.split("\n")
headers = lines[0].split(",")

# Zbuduj słownik dla jednego produktu
for line in lines[1:]:
    if line.strip() == "":
        continue

    values = line.split(",")

    product = {}

    for i in range(len(headers)):
        if headers[i] == "price":
            product[headers[i]] = int(values[i])
        else:
            product[headers[i]] = values[i]

    print(product)

# Teraz dodaj każdy słownik do listy products
products = []

for line in lines[1:]:
    if line.strip() == "":
        continue

    values = line.split(",")

    product = {}

    for i in range(len(headers)):
        if headers[i] == "price":
            product[headers[i]] = int(values[i])
        else:
            product[headers[i]] = values[i]

    products.append(product)

print(products)

# Policz sumę cen produktów droższych niż 10
total = 0
count = 0

for product in products:
    if product["price"] > 10:
        total = total + product["price"]
        count = count + 1

print("Suma:", total)
print("Liczba produktów:", count)

# Policz średnią
average = total / count
print("Średnia cena produktów droższych niż 10:", average)
```
