## 1. Filtrowanie: wybierz liczby parzyste

**Wskazówka:** Parzysta: x % 2 == 0


```python
nums = [1, 2, 3, 4, 5, 6, 7, 8]   # Tworzymy listę liczb od 1 do 8.

evens = []                        # Tworzymy pustą listę na liczby parzyste.

for num in nums:                  # Przechodzimy kolejno przez każdy element listy nums.
    if num % 2 == 0:              # Sprawdzamy, czy dana liczba jest parzysta.
        evens.append(num)         # Jeśli tak, dodajemy ją do listy evens.


print(f"Evens: {evens}")
```

```python
try:
    assert evens == [2,4,6,8]
    print('OK ✅')
except NameError:
    print('Zmienna evens nie istnieje (sprawdź nazwę)')
except Exception as e:
    print('Błąd:', e)
```

## 2. Słownik: zlicz ile razy występuje każde słowo

**Wskazówka:** Jeśli klucz nie istnieje: ustaw 0, potem zwiększ.


```python
words = ['cat', 'dog', 'cat', 'mouse', 'dog', 'cat']   # Tworzymy listę słów, w której część elementów się powtarza.

counts = {}                                             # Tworzymy pusty słownik, w którym będziemy liczyć wystąpienia słów.

for word in words:                                      # Przechodzimy po kolei przez każde słowo z listy words.
    if word in counts:                                  # Sprawdzamy, czy dane słowo już istnieje jako klucz w słowniku.
        counts[word] += 1                               # Jeśli tak, zwiększamy jego licznik o 1.
    else:                                               # Jeśli słowa jeszcze nie ma w słowniku...
        counts[word] = 1                                # ...dodajemy je do słownika i ustawiamy licznik na 1.

print(f"Counts: {counts}")
```

```python
try:
    assert counts == {'cat':3,'dog':2,'mouse':1}
    print('OK ✅')
except Exception as e:
    print('Błąd:', e)
```

## 3. Średnia bez NumPy

**Wskazówka:** mean = suma / liczba_elementów


```python
nums = [2, 4, 4, 4, 5, 5, 7, 9]   # Tworzymy listę liczb, z których chcemy obliczyć średnią.

total = 0                         # Ustawiamy zmienną total na 0; będzie przechowywać sumę wszystkich liczb.

for num in nums:                  # Przechodzimy po kolei przez każdy element listy nums.
    total += num                  # Dodajemy bieżącą liczbę do sumy total.

mean = total / len(nums)          # Dzielimy sumę wszystkich liczb przez liczbę elementów listy.

print(f"Mean: {mean}")
```

```python
try:
    assert abs(mean - 5.0) < 1e-9
    print('OK ✅')
except Exception as e:
    print('Błąd:', e)
```

## 4. Miara rozrzutu (intuicyjnie): rozstęp = max - min

**Wskazówka:** Użyj wbudowanych max(...) i min(...).


```python
nums = [2, 4, 4, 4, 5, 5, 7, 9]   # Tworzymy listę liczb, dla których chcemy obliczyć rozstęp.

data_range = max(nums) - min(nums) # Obliczamy rozstęp danych: największa liczba minus najmniejsza liczba.
print(f"Data range: {data_range}")
```

```python
try:
    assert data_range == 7
    print('OK ✅')
except Exception as e:
    print('Błąd:', e)
```

## 5. Mini-dataset: transakcje
Zobaczysz listę rekordów (listę słowników). To jest bardzo typowy format „przed pandas”.

```python
transactions = [
    {'city': 'lublin',   'amount': 120},   # Pierwsza transakcja: miasto Lublin, kwota 120.
    {'city': 'warszawa', 'amount':  80},   # Druga transakcja: miasto Warszawa, kwota 80.
    {'city': 'warszawa', 'amount': 200},   # Trzecia transakcja: miasto Warszawa, kwota 200.
    {'city': 'lublin',   'amount':  50},   # Czwarta transakcja: miasto Lublin, kwota 50.
]

sum_warszawa = 0                           # Ustawiamy początkową sumę dla Warszawy na 0.
count_ge100 = 0                            # Ustawiamy licznik transakcji o kwocie >= 100 na 0.
share_ge100 = None                         # Tworzymy zmienną na procent takich transakcji; na razie brak wyniku.

for transaction in transactions:           # Przechodzimy po kolei przez każdą transakcję z listy.
    if transaction['city'] == 'warszawa':  # Sprawdzamy, czy miasto w danej transakcji to Warszawa.
        sum_warszawa += transaction['amount']  # Jeśli tak, dodajemy kwotę tej transakcji do sumy dla Warszawy.

    if transaction['amount'] >= 100:       # Sprawdzamy, czy kwota transakcji jest większa lub równa 100.
        count_ge100 += 1                   # Jeśli tak, zwiększamy licznik takich transakcji o 1.

share_ge100 = (count_ge100 / len(transactions)) * 100  # Obliczamy procent transakcji z kwotą >= 100.

print(sum_warszawa, count_ge100, share_ge100)  # Wypisujemy obliczone wyniki.
```

```python
try:
    assert sum_warszawa == 280
    assert count_ge100 == 2
    assert abs(share_ge100 - 50.0) < 1e-9
    print('OK ✅')
except Exception as e:
    print('Błąd:', e)
```
